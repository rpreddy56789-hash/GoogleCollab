```python
import pandas as pd
import numpy as np
```

```python
url = "/content/sample_data/deliveries.csv"
```

```python
df = pd.read_csv(url)
```

```python
print(df.dtypes)
```

```python
print('Current DataFrame dtypes:')
print(df.dtypes)
```

```python
# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'], format='%b %d, %Y', errors='coerce')

# Convert categorical columns to 'category' type
categorical_cols = ['stage', 'venue', 'batting_team', 'bowling_team', 'wicket_type', 'striker', 'bowler', 'player_dismissed', 'fielder']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')

print('\nDataFrame dtypes after type conversion:')
print(df.dtypes)

print('\nDisplaying the first 5 rows of the DataFrame with updated types:')
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))
sns.boxplot(x='batting_team', y='runs_of_bat', data=df)
plt.title('Distribution of Runs Scored per Ball by Batting Team')
plt.xlabel('Batting Team')
plt.ylabel('Runs Scored per Ball')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
